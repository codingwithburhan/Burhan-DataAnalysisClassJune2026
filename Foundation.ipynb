```python
choice = input("Do you want Coffee or Tea? ")

if choice == "Coffee":
    milk = input("Do you want to add Milk? ")
    if milk == "Yes":
        print("Congratulations! Coffee has been added to your order with milk.")
    else:
        print("Congratulations! Coffee has been added to your order without milk.")

elif choice == "Tea":
    sugar = input("Do you want to add sugar? ")

    if sugar == "yes":
        print("Congratulations! Tea has been added to your order with sugar.")
    else:
        print("Congratulations! Tea has been added to your order without sugar.")

else:
    print("Invalid choice.")
```
